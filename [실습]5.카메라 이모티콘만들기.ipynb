```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
print("üå´üõ∏")
```

```python
import os
my_image_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/image.jpg'
```

```python
img_bgr = cv2.imread(my_image_path)    #- OpenCVÎ°ú Ïù¥ÎØ∏ÏßÄÎ•º ÏùΩÏñ¥ÏÑú
img_bgr = cv2.resize(img_bgr, (640, 360))    # 640x360Ïùò ÌÅ¨Í∏∞Î°ú Resize
img_show = img_bgr.copy()      #- Ï∂úÎ†•Ïö© Ïù¥ÎØ∏ÏßÄ Î≥ÑÎèÑ Î≥¥Í¥Ä
plt.imshow(img_bgr)
plt.show()
##ÏñºÍµ¥ ÏÉâÏóê ÏôúÏù¥Î†áÍ≤å Ìë∏Î•∏ÎπõÏù¥ Í∞êÎèÑÎäî Í±∏ÍπåÏöî?
##Ïù¥Í≤ÉÏùÄ opencvÏùò ÌäπÏßï ÎïåÎ¨∏ÏûÖÎãàÎã§. 
```

```python
# plt.imshow Ïù¥Ï†ÑÏóê RGB Ïù¥ÎØ∏ÏßÄÎ°ú Î∞îÍæ∏Îäî Í≤ÉÏùÑ ÏûäÏßÄÎßàÏÑ∏Ïöî. 
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

## ÏñºÍµ¥ÏúÑÏπòÏù∏Ïãù

```python
##dlib_hot detector
import dlib
detector_hog = dlib.get_frontal_face_detector()   #- detector ÏÑ†Ïñ∏
print("üå´üõ∏")
```

```python
##detector_ÏñºÍµ¥Ïùòbounding boxÏ∂îÏ∂ú
##cvtcolorÎ•º ÌôúÏö©Ìï¥ rgbÎ°ú Ï†ÑÌôò
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
dlib_rects = detector_hog(img_rgb, 1)   #- (image, num of img pyramid)
print("üå´üõ∏")
```

```python
##Ï∞æÏùÄ ÏñºÍµ¥ ÌôîÎ©¥Ïóê Ï∂úÎ†•
print(dlib_rects)   # Ï∞æÏùÄ ÏñºÍµ¥ÏòÅÏó≠ Ï¢åÌëú

for dlib_rect in dlib_rects:
    l = dlib_rect.left()
    t = dlib_rect.top()
    r = dlib_rect.right()
    b = dlib_rect.bottom()

    cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA)

img_show_rgb =  cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

dlib detector Îäî dlib.rectangles ÌÉÄÏûÖÏùò Í∞ùÏ≤¥Î•º Î∞òÌôòÌï©ÎãàÎã§.

## ÎûúÎìúÎßàÌÅ¨

```python
import os
model_path = os.getenv('HOME')+'/aiffel/camera_sticker/models/shape_predictor_68_face_landmarks.dat'
landmark_predictor = dlib.shape_predictor(model_path)
print("üå´üõ∏")
##landmark_predictor Îäî RGB Ïù¥ÎØ∏ÏßÄÏôÄ
##dlib.rectangleÏùÑ ÏûÖÎ†• Î∞õÏïÑ dlib.full_object_detection Î•º Î∞òÌôòÌï©ÎãàÎã§.
```

```python
list_landmarks = []
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
    list_landmarks.append(list_points)

print(len(list_landmarks[0]))
```

```python
##ÌôïÏù∏
for landmark in list_landmarks:
    for idx, point in enumerate(list_points):
        cv2.circle(img_show, point, 2, (0, 255, 255), -1) # yellow

img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

## Ïä§Ìã∞Ïª§Ï†ÅÏö©

```python
##ÎààÏΩîÏûÖÎì±Ïùò Ï¢åÌëúÌôïÏù∏
for dlib_rect, landmark in zip(dlib_rects, list_landmarks):
    print (landmark[30]) # nose center index : 30
    x = landmark[30][0]
    y = landmark[30][1] - dlib_rect.width()//2
    w = dlib_rect.width()
    h = dlib_rect.width()
    print ('(x,y) : (%d,%d)'%(x,y))
    print ('(w,h) : (%d,%d)'%(w,h))
```

```python
##ÏÇ¨ÏßÑÏóê ÎßûÍ≤å Ïù¥Î™®Ìã∞ÏΩò resize
import os
sticker_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/king.png'
img_sticker = cv2.imread(sticker_path)
img_sticker = cv2.resize(img_sticker, (w,h))
print (img_sticker.shape)
```

```python
refined_x = x - w // 2  # left
refined_y = y - h       # top
print ('(x,y) : (%d,%d)'%(refined_x, refined_y))
```

```python
##Ïù¥ÎØ∏ÏßÄÌÅ¨Í∏∞Ïóê ÎßûÍ≤å resize
img_sticker = img_sticker[-refined_y:]
print (img_sticker.shape)
```

```python
refined_y = 0
print ('(x,y) : (%d,%d)'%(refined_x, refined_y))
```

```python
sticker_area = img_show[refined_y:img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
img_show[refined_y:img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==0,sticker_area,img_sticker).astype(np.uint8)
```

```python
plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
plt.show()
```

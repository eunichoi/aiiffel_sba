```python
##ìŠ¤ë ˆë“œìƒì„±
from threading import *
from time import sleep

Stopped = False

def worker(work, sleep_sec):    # ì¼ê¾¼ ìŠ¤ë ˆë“œì…ë‹ˆë‹¤.
    while not Stopped:   # ê·¸ë§Œ í•˜ë¼ê³  í• ë•Œê¹Œì§€
        print('do ', work)    # ì‹œí‚¤ëŠ” ì¼ì„ í•˜ê³ 
        sleep(sleep_sec)    # ì ê¹ ì‰½ë‹ˆë‹¤.
    print('retired..')           # ì–¸ì  ê°€ ì´ êµ´ë ˆë¥¼ ë²—ì–´ë‚˜ë©´, ì€í‡´í•  ë•Œê°€ ì˜¤ê² ì§€ìš”?
        
t = Thread(target=worker, args=('Overwork', 3))    # ì¼ê¾¼ ìŠ¤ë ˆë“œë¥¼ í•˜ë‚˜ ìƒì„±í•©ë‹ˆë‹¤. ì—´ì‹¬íˆ ì¼í•˜ê³  3ì´ˆê°„ ì‰½ë‹ˆë‹¤.
t.start()    # ì¼ê¾¼, ì´ì œ ì¼ì„ í•´ì•¼ì§€? ğŸ˜ˆ
```

```python
##ìˆ«ìë“¤ì´ ì†Œìˆ˜ì¸ì§€ ì•„ë‹Œì§€ íŒë³„
import math
import concurrent

PRIMES = [
    112272535095293,
    112582705942171,
    112272535095293,
    115280095190773,
    115797848077099,
    1099726899285419]

print("ğŸª    .\nÂ·   *\n  *   *\nğŸŒ² ğŸ¦• ğŸŒ³")

###---

##ì†Œìˆ˜íŒë³„í•¨ìˆ˜
def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False

    sqrt_n = int(math.floor(math.sqrt(n)))
    for i in range(3, sqrt_n + 1, 2):
        if n % i == 0:
            return False
    return True


###---
##ì†Œìˆ˜íŒë³„í•¨ìˆ˜í˜¸ì¶œ
def main():
    with concurrent.futures.ProcessPoolExecutor() as executor:
        for number, prime in zip(PRIMES, executor.map(is_prime, PRIMES)):
            print('%d is prime: %s' % (number, prime))

if __name__ == '__main__':
    main()
```

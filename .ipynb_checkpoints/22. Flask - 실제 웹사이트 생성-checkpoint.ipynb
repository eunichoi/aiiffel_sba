```python
##Ïã§ÌñâÏãúÍ∞Ñ ÌôïÏù∏
import datetime

# ÏΩîÎìú ÏÑ†Ïñ∏Î∂Ä
def main_function():     
     print ("test function")

# ÏΩîÎìú Ïã§ÌñâÎ∂Ä   
print (datetime.datetime.now()) #- ÌÖçÏä§Ìä∏ ÌëúÏãú Ï†Ñ ÏãúÍ∞Ñ
main_function()
print (datetime.datetime.now()) #- ÌÖçÏä§Ìä∏ ÌëúÏãú ÌõÑ ÏãúÍ∞Ñ
```

```python
##Ïã§ÌñâÏãúÍ∞ÑÍ¥ÄÎ†® Îç∞ÏΩîÎ†àÏù¥ÌÑ∞ ÏÉùÏÑ±
# ÏΩîÎìú ÏÑ†Ïñ∏Î∂Ä
def datetime_decorator(func):
        def decorated():
                print(datetime.datetime.now())
                func()
                print(datetime.datetime.now())
        return decorated

@datetime_decorator
def main_function():
        print ("test function")


# ÏΩîÎìú Ïã§ÌñâÎ∂Ä  
main_function()
```

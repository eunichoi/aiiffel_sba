```python
my_list = ['a','b','c','d']

# ì¸ìë¡œ ë°›ì€ ë¦¬ìŠ¤íŠ¸ë¡œë¶€í„° ë°ì´í„°ë¥¼ í•˜ë‚˜ì”© ê°€ì ¸ì˜¤ëŠ” ì œë„ˆë ˆì´í„°ë¥¼ ë¦¬í„´í•˜ëŠ” í•¨ìˆ˜
def get_dataset_generator(my_list):
    result_list = []
    for i in range(2):
        for j in my_list:
            yield (i, j)
            print('>>  1 data loaded..')

dataset_generator = get_dataset_generator(my_list)
for X, y in dataset_generator:
    print(X, y)
    
```
